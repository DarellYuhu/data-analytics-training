```python
orders = []

for i in range(num_orders):

    order_id = i + 1

    customer_id = random.choice(
        customers["customer_id"].tolist()
    )

    customer = customers[
        customers["customer_id"] == customer_id
    ].iloc[0]

    acquisition_date = pd.to_datetime(
        customer["acquisition_date"]
    )

    today = datetime.today()

    days_since_signup = (
        today - acquisition_date
    ).days

    order_date = (
        acquisition_date
        + timedelta(days=random.randint(0, days_since_signup))
    ).date()

    status = random.choices(
        order_statuses,
        weights=status_weights,
        k=1
    )[0]

    payment_method = random.choices(
        payment_methods,
        weights=payment_weights,
        k=1
    )[0]

    order = {
        "order_id": order_id,
        "customer_id": customer_id,
        "order_date": order_date,
        "order_status": status,
        "payment_method": payment_method
    }

    orders.append(order)
```

```python
df_orders = pd.DataFrame(orders)
```

```python
order_items = []
num_order_items = 0

for _, order in df_orders.iterrows():

    # Number of different products in this order
    num_products = random.randint(1, 5)

    # Randomly select products
    selected_products = products.sample(num_products)

    # Loop through each selected product
    for _, product in selected_products.iterrows():

        quantity = random.randint(1, 3)

        num_order_items += 1

        order_item = {
            "order_item_id": num_order_items,
            "order_id": order["order_id"],
            "product_id": product["ProductID"],
            "quantity": quantity,
            "unit_price": product["UnitPrice"]      # We'll verify this column name next
        }

        order_items.append(order_item)

# Create the DataFrame AFTER the loops
df_order_items = pd.DataFrame(order_items)
```

```python
import pandas as pd
import random
from datetime import datetime, timedelta

# ===========================
# Marketing Spend Generation
# ===========================

marketing = []

channels = [
    "Google",
    "Facebook",
    "TikTok",
    "Email",
    "Organic"
]

start_date = datetime(2024, 1, 1)
end_date = datetime(2025, 12, 31)

current_date = start_date

while current_date <= end_date:

    for channel in channels:

        spend = round(random.uniform(100, 1000), 2)

        marketing.append({
            "date": current_date.date(),
            "channel": channel,
            "spend": spend
        })

    current_date += timedelta(days=1)

# Convert to DataFrame
df_marketing = pd.DataFrame(marketing)

# Preview
print("Rows generated:", len(df_marketing))
display(df_marketing.head())

# Save to CSV
df_marketing.to_csv("marketing_spend.csv", index=False)

print("marketing_spend.csv has been created successfully!")

```
